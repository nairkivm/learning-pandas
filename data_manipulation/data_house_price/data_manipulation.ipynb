# <font color='white'> 1. Study Case 1: House Pricing
- Diberikan dua buah data :
    - `house-pricing`
    - `house-spesification`

```python
import pandas as pd

housePriceDF = pd.read_csv("house-price.csv")
housePriceDF.info()
```

```python
houseSpecDF = pd.read_csv("house-specification.csv")
houseSpecDF.info()
```

- Gabungkan kedua data tersebut.
- Gabung dengan menggunakan method `.join()`
- Gabung berdasarkan :
    - Kolom `house_id`
- Ketentuan :
    - Jenis join yang dilakukan **inner join**
    - **Left table:**
        - `house_spesifications`
    - **Right table:**
        - `house_pricing`

```python
# Set house_id as index
housePriceDF = housePriceDF.set_index("house_id")
houseSpecDF = houseSpecDF.set_index("house_id")
```

```python
# Join the tables (the result is assigned to "housePriceSpecDF")
houseDF = houseSpecDF.join(housePriceDF, how="inner")
```

```python
houseDF.head()
```

- Dari data hasil join, anda diminta melakukan **Aggregation** untuk mencari :
    - **price terendah**
    - **price tertinggi**
    - **rata-rata price**
    - Notes :
        - **Bulatkan dua angka dibelakang koma**

```python
# Cari price terendah
def getLowestPrice(df):
    lowestPrice = min(df["Price"])
    print(f"Harga rumah terendah adalah : ${lowestPrice:.2f}")
    return lowestPrice

lowestPrice = getLowestPrice(houseDF)
```

```python
# Cari price tertinggi
def getHighestPrice(df):
    highestPrice = max(df["Price"])
    print(f"Harga rumah tertinggi adalah : ${highestPrice:.2f}")
    return highestPrice

highestPrice = getHighestPrice(houseDF)
```

```python
# Cari rata-rata price
def getAveragePrice(df):
    averagePrice = (df["Price"].mean())
    print(f"Harga rumah rata-rata adalah : ${averagePrice:.2f}")
    return averagePrice

averagePrice = getAveragePrice(houseDF)
```

- Tampilkan baris :
    - **price terendah**
    - **price tertinggi**
    - **price dibawah rata-rata**
    - **price diatas rata-rata**

```python
# Cari price terendah
def getLowestPriceRows(df):
    lowestPrice = min(df["Price"])
    lowestPriceRow = df[df["Price"] == lowestPrice]
    return lowestPriceRow

getLowestPriceRows(houseDF)
```

```python
# Cari price tertinggi
def getHighestPriceRows(df):
    highestPrice = max(df["Price"])
    highestPriceRow = df[df["Price"] == highestPrice]
    return highestPriceRow

getHighestPriceRows(houseDF)
```

```python
# Cari data di bawah rata-rata price
def getUnderAveragePriceRows(df):
    averagePrice = (df["Price"].mean())
    underAveragePriceRows = df[df["Price"] < averagePrice]
    return underAveragePriceRows

getUnderAveragePriceRows(houseDF)
```

```python
# Cari data di atas rata-rata price
def getAboveAveragePriceRows(df):
    averagePrice = (df["Price"].mean())
    aboveAveragePriceRows = df[df["Price"] > averagePrice]
    return aboveAveragePriceRows

getAboveAveragePriceRows(houseDF)
```

- **Buat data binning** pada dataframe tersebut (disimpan pada kolom 'house_category').
- Buat data binning **berdasarkan kolom SquareFeet.**
- Kelompokkan data tersebut dengan **ketentuan berikut :**
    - **`1000 - 1500` : `Small Homes`**
    - **`1501 - 2000` : `Medium Homes`**
    - **`2001 - 2500` : `Midsize Homes`**
    - **`2501 - 3000` : `Large Homes`**
- Tampilkan jumlah data dari setiap house_category.

```python
SQUARE_FEET_BINS = [1000, 1500, 2000, 2500, 3000]
SQUARE_FEET_CATEGORIES = ["Small Homes", "Medium Homes", "Midsize Homes", "Large Homes"]
houseDF["house_category"] = (
    pd.cut(
        houseDF["SquareFeet"],
        bins=SQUARE_FEET_BINS,
        labels=SQUARE_FEET_CATEGORIES
    )
)
houseDF.head()
```

- Melakukan Grouping berdasarkan kolom `house_category`.

```python
GROUP_COLUMNS = ["house_category"]
houseDFGrouped = houseDF.groupby(GROUP_COLUMNS)
houseDFGrouped
```

- Dari data hasil grouping :
    - Tampilkan 5 data pertama pada masing-masing group


```python
# 5 data pertama untuk berbagai kategori
houseCategoryDFs = {}
for ctg in SQUARE_FEET_CATEGORIES:
    houseCategoryDFs[ctg] = (
        houseDFGrouped
        .get_group(ctg)
    )

SELECTED_CATEGORY = "Large Homes" 
houseCategoryDFs[SELECTED_CATEGORY].head(5)
```

- **Pada house_category Small Homes dan Medium Homes:**
    - **Tampilkan data** atau baris yang mempunyai nilai pada kolom Price :
        - `min`
        - `max`


```python
def getSelectedDF(dataFrame, priceValue):
    selectedDF = (
        dataFrame[
            (dataFrame["Price"] == priceValue)
        ]
    )
    return selectedDF

# Select data that contains min value of Price from "Small Homes" category
SELECTED_CATEGORY = "Small Homes" 
MIN_PRICE = houseCategoryDFs[SELECTED_CATEGORY]["Price"].min()

selectedDF = getSelectedDF(houseCategoryDFs[SELECTED_CATEGORY], MIN_PRICE)

selectedDF
```

```python
# Select data that contains max value of Price from "Small Homes" category
SELECTED_CATEGORY = "Small Homes" 
MAX_PRICE = houseCategoryDFs[SELECTED_CATEGORY]["Price"].max()

selectedDF = getSelectedDF(houseCategoryDFs[SELECTED_CATEGORY], MAX_PRICE)

selectedDF
```

```python
# Select data that contains min value of Price from "Medium Homes" category
SELECTED_CATEGORY = "Medium Homes" 
MIN_PRICE = houseCategoryDFs[SELECTED_CATEGORY]["Price"].min()

selectedDF = getSelectedDF(houseCategoryDFs[SELECTED_CATEGORY], MIN_PRICE)

selectedDF
```

```python
# Select data that contains max value of Price from "Medium Homes" category
SELECTED_CATEGORY = "Medium Homes" 
MAX_PRICE = houseCategoryDFs[SELECTED_CATEGORY]["Price"].max()

selectedDF = getSelectedDF(houseCategoryDFs[SELECTED_CATEGORY], MAX_PRICE)

selectedDF
```
