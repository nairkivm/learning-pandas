```python
import pandas as pd

rawPollutedCountriesDF_1 = pd.read_csv("most-polluted-countries-1.csv")
rawPollutedCountriesDF_1.info()
```

```python
rawPollutedCountriesDF_2 = pd.read_csv("most-polluted-countries-2.csv")
rawPollutedCountriesDF_2.info()
```

```python
# Preview data
rawPollutedCountriesDF_2.head()
```

- **Gabungkan kedua data tersebut.**
- Gabung dengan menggunakan method `.concat()`
- Penggabungan yang dilakukan adalah **penggabungan baris.**
- Ketentuan :
    - **Top table:**
        - `mp_country_1`
    - **Bottom table:**
        - `mp_country_2`
- Notes :
    - Buat indexnya menjadi continue.

```python
# Join the dataframes
pollutedCountriesDF = pd.concat(
    [
        rawPollutedCountriesDF_1,
        rawPollutedCountriesDF_2
    ],
    axis=0,
    ignore_index=True
)

# Preview the data
pollutedCountriesDF.head()
```

```python
# Preview the data
pollutedCountriesDF.tail()
```

- Melakukan **pengelompokkan** data berdasarkan kolom `country_region`.
- Menampilkan data pertama dari setiap region

```python
# Group the dataframe by "country_region" column
pollutedCountriesDFGrouped = pollutedCountriesDF.groupby("country_region")
# Get first data from each group
selectedPollutedCountriesDF = (
    pollutedCountriesDFGrouped
    .first()
)
# Preview the dataframe
selectedPollutedCountriesDF
```

- Mencari Negara dengan polusi (`pollution_2023`) **tertinggi dan terendah pada region Asia.**
- **Print Negara dengan polusi tertinggi dan terendah.**
- **Tampilkan baris negara tersebut.**

```python
# Select dataframe by group
selectedPollutedCountriesDF = (
    pollutedCountriesDFGrouped
    .get_group("Asia")
)
def getMostPollutedCountry(df):
    mostPollution = max(df["pollution_2023"])
    mostPollutedRecord = (
        df[df["pollution_2023"] == mostPollution]
    )
    mostPollutedCountry = (
        mostPollutedRecord
        ["country_name"]
        .values[0]
    )
    print(f"Negara dengan polusi tertinggi : {mostPollutedCountry}")
    return mostPollutedRecord

mostPollutedRecord = getMostPollutedCountry(selectedPollutedCountriesDF)
mostPollutedRecord
```

```python
def getLeastPollutedCountry(df):
    leastPollution = min(df["pollution_2023"])
    leastPollutedRecord = (
        df[df["pollution_2023"] == leastPollution]
    )
    leastPollutedCountry = (
        leastPollutedRecord
        ["country_name"]
        .values[0]
    )
    print(f"Negara dengan polusi terendah : {leastPollutedCountry}")
    return leastPollutedRecord

leastPollutedRecord = getLeastPollutedCountry(selectedPollutedCountriesDF)
leastPollutedRecord
```

- Mencari negara dengan polusi (`pollution_2023`) **dibawah dan diatas rata-rata.**
- **Tampilkan baris negara-negara tersebut.**

```python
meanPollution = pollutedCountriesDF["pollution_2023"].mean()

# Show under average pollution countries

underAveragePollutedCountriesDF = (
    pollutedCountriesDF[
        pollutedCountriesDF["pollution_2023"] < meanPollution
    ]
)
underAveragePollutedCountriesDF
```

```python
# Show above average pollution countries

aboveAveragePollutedCountriesDF = (
    pollutedCountriesDF[
        pollutedCountriesDF["pollution_2023"] > meanPollution
    ]
)
aboveAveragePollutedCountriesDF
```
