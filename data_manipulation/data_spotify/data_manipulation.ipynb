- Diberikan dua buah data :
    - `spotify-track`
    - `spotify-album`

```python
import pandas as pd

rawTrackDF = pd.read_csv("spotify-track.csv")
rawTrackDF.info()
```

```python
rawAlbumDF = pd.read_csv("spotify-album.csv")
rawAlbumDF.info()
```

```python
# Preview data
rawAlbumDF.head()
```

- Gabungkan kedua data tersebut.
- Gabung dengan menggunakan method `.merge()`
- Gabung berdasarkan :
    - Pada data `spotify_tracks` :
        - Kolom `track_id`
    - Pada data `spotify_albums` :
        - Kolom `id_track`
- Ketentuan :
    - Jenis join yang dilakukan **inner join**
    - **Left table:**
        - `spotify_tracks`
    - **Right table:**
        - `spotify_albums`
- Notes :
    - Pastikan tidak ada data `track_id` yang duplikat
    - Setelah dilakukan merge, hapus kolom `id_track`

```python
# Join the dataframes
trackAlbumDF = pd.merge(
    rawTrackDF,
    rawAlbumDF,
    how="inner",
    left_on="track_id",
    right_on="id_track"
)

# Delete duplicated data on "track_id" column
trackAlbumDF = (
    trackAlbumDF
    .drop_duplicates(
        subset="track_id", 
        keep="last",
        ignore_index=True
    )
)

# Delete "id_track" column
trackAlbumDF = (
    trackAlbumDF.drop(
        columns="id_track"
    )
)

# Preview the data
trackAlbumDF.head()
```

- Melakukan **ekstraksi** data pada kolom `track_album_release_date`. Ekstrak data :
    - `tahun`
    - `bulan`
    - `tanggal`
- Ekstraksi kolom tersebut akan **disimpan pada kolom :**
    - `year_album`
    - `month_album`
    - `day_album`

```python
# Format "track_album_release_date" column as datetime
trackAlbumDF["track_album_release_date"] = (
    pd.to_datetime(
        trackAlbumDF["track_album_release_date"]
    )
)

# Extract "track_album_release_date" column
## Extract "year"
trackAlbumDF["year_album"] = trackAlbumDF["track_album_release_date"].dt.year 
## Extract "month"
trackAlbumDF["month_album"] = trackAlbumDF["track_album_release_date"].dt.month 
## Extract "date"
trackAlbumDF["day_album"] = trackAlbumDF["track_album_release_date"].dt.date 


trackAlbumDF.info()
```

```python
# Preview the data
trackAlbumDF.head()
```

- Melakukan **aggregasi** untuk :
    - **Memprint** `year_album` dengan `track_popularity` paling tingggi
    - **Menampilkan baris** atau data dengan `track_popularity` paling tinggi (maksimal).

```python
# Find the "year_album" of the track with "highest popularity"
def getAggregatedValue(df):
    highestPopularity = max(df["track_popularity"])
    yearAlbumOfTheTrack = (
        df
        [df["track_popularity"] == highestPopularity]
        ["year_album"]
        .values[0]
    )
    print(f"Track popularity paling tinggi terdapat pada tahun : {yearAlbumOfTheTrack}")
    return yearAlbumOfTheTrack

yearAlbumOfTheTrack = getAggregatedValue(trackAlbumDF)
```

```python
# Show the record of the track with "highest popularity"
def getAggregatedRecord(df):
    highestPopularity = max(df["track_popularity"])
    record = (
        df[df["track_popularity"] == highestPopularity]
    )
    return record

getAggregatedRecord(trackAlbumDF)
```

- **Mengelompokkan data** berdasarkan kolom `year_album`
- Menampilkan data dengan `track_popularity` paling tinggi pada `year_album` :
    - 2020 sampai 2016

```python
# Group the dataframe based on "year_album"
trackAlbumDFGrouped = (
    trackAlbumDF.groupby("year_album")
)
# Get all index which the data contains highest "track_popularity" in its corresponding group
selectedIndexSeries = (
    trackAlbumDFGrouped
    ["track_popularity"]
    .idxmax()
)
# Get the dataframe with highest "track_popularity"
filteredDF = trackAlbumDF.loc[selectedIndexSeries]
# Sort the dataframe by descending order of "year_album"
sortedDF = filteredDF.sort_values(by="year_album",ascending=False)
# Select only data from 2016-2020
YEAR_RANGE = range(2016, 2020 + 1)
selectedDF = sortedDF[sortedDF["year_album"].isin(YEAR_RANGE)]
selectedDF
```
