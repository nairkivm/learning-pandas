# <font color='white'> 4. Study Case 4: Youtube Videos
- Diberikan dua buah data :
    - `youtube-videos-1.csv`
    - `youtube-videos-2.csv`
- Data tersebut merupakan data pada tahun 2023.

```python
import pandas as pd

rawYoutubeDF_1 = pd.read_csv("youtube-videos-1.csv")
rawYoutubeDF_1.info()
```

```python
# Preview data
rawYoutubeDF_1.head()
```

```python
rawYoutubeDF_2 = pd.read_csv("youtube-videos-2.csv")
rawYoutubeDF_2.info()
```

```python
# Preview data
rawYoutubeDF_2.head()
```

- **Gabungkan kedua data tersebut.**
- Gabung dengan menggunakan method `.concat()`
- Penggabungan yang dilakukan adalah **penggabungan baris.**
- Ketentuan :
    - **Top table:**
        - `yt_videos_1`
    - **Bottom table:**
        - `yt_videos_2`
- Notes :
    - Buat indexnya menjadi continue.

```python
# Join the dataframes
youtubeDF = pd.concat(
    [
        rawYoutubeDF_1,
        rawYoutubeDF_2
    ],
    axis=0,
    ignore_index=True
)

# Preview the data
youtubeDF.head()
```

- Sebelum melakukan pengelompokkan, lakukan :
    - **Casting** tipe data dari kolom `published_at` menjadi datetime.
    - Ekstrak kolom `published_at`
    - **Buat kolom baru** untuk menyimpan data hasil ekstraksi

```python
# Format "published_at" column as datetime
youtubeDF["published_at"] = pd.to_datetime(youtubeDF["published_at"])
# Extract "published_at" column
## Extract year
youtubeDF["year_published"] = youtubeDF["published_at"].dt.year
## Extract month
youtubeDF["month_published"] = youtubeDF["published_at"].dt.month
## Extract date
youtubeDF["day_published"] = youtubeDF["published_at"].dt.date
## Delete published_at
youtubeDF = youtubeDF.drop(columns="published_at")
# Preview the dataframe
youtubeDF.head()
```

- **Lakukan pengelompokkan data** berdasarkan kolom `month_published`

```python
# Group the dataframe based on "month_published"
youtubeDFGrouped = youtubeDF.groupby("month_published", sort=True)
```

- Menampilkan **urutan data dengan jumlah view terbanyak pada tahun 2023.** berdasarkan pengelompokan tersebut

```python
# Get indexes of records with highest view_count in each group
highestViewCountIndexes = (
    youtubeDFGrouped
    ["view_count"]
    .idxmax()
)
# Get all records that match the filter
filteredYoutubeDF = youtubeDF.loc[highestViewCountIndexes]
filteredYoutubeDF
```

- Menampikan **urutan data dengan jumlah view terbanyak pada bulan februari tahun 2023.**

```python
# Group the dataframe based on "year_album"
febYoutubeDF = (
    youtubeDFGrouped.get_group(2)
)
# Sort the dataframe
sortedFebYoutubeDF = (
    febYoutubeDF
    .sort_values(
        by="view_count", 
        ascending=False,
        ignore_index=True
    )
)
sortedFebYoutubeDF
```
