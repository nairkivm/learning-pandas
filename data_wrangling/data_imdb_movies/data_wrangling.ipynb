```python
# Import packages
import pandas as pd
```

- Read `IMDB_Movies.csv` dataset
- Data tersebut berisi tentang informasi mengenai film - film yang ada di website [IMDB](https://www.imdb.com/)

```python
rawMovieDataFrame = pd.read_csv('data/IMDB_Movies.csv')
rawMovieDataFrame
```

- Cek shape dari data yang dimiliki

```python
rawMovieDataFrameShape = rawMovieDataFrame.shape
rawMovieDataFrameShape
```

- Pilihlah kolom - kolom berikut:
    - `movie_title`
    - `title_year`
    - `genres`
    - `duration`
    - `director_name`
    - `imdb_score`

```python
movieColumns = [
    'movie_title',
    'title_year',
    'genres',
    'duration',
    'director_name',
    'imdb_score'
]
movieDataFrame = rawMovieDataFrame[movieColumns]
movieDataFrame
```

- Setelah itu, lakukan proses filter data yang dimana memiliki skor IMDB lebih dari sama dengan `6.5`

```python
filteredMovieDataFrame = (
    movieDataFrame[
        movieDataFrame['imdb_score'] >= 6.5
    ]
)
filteredMovieDataFrame
```
