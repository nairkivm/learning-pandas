```python
# Import packages
import pandas as pd
from sqlalchemy import create_engine
```

```python
# create connection with postgres
conn = create_engine("postgresql://postgres:auliya01@localhost/pagila")
```

```python
query = "SELECT * FROM film"

rentalDataFrame = pd.read_sql(sql = query, con = conn)

rentalDataFrame
```

- Langkah selanjutnya adalah select kolom - kolom berikut:
    - `film_id`
    - `title`
    - `release_year`
    - `rental_rate`
    - `length`
    - `rating`
    - `last_update`

```python
rentalDataFrameColumns = [
    'film_id', 
    'title',
    'release_year',
    'rental_rate',
    'length',
    'rating',
    'last_update'
]

rentalDataFrame = rentalDataFrame[rentalDataFrameColumns]

rentalDataFrame
```

- Sebelum melakukan filter data, coba cek bagaimana value yang ada di kolom `rating`

```python
rentalDataFrame['rating'].value_counts()
```

- Berdasarkan data tersebut, lakukanlah filter data yang dimana rating memiliki value `PG` dan memiliki durasi yang lebih dari sama dengan `90`

```python
filteredRentalDataFrame = (
    rentalDataFrame[
        (rentalDataFrame['rating'] == 'PG') & 
        (rentalDataFrame['length'] >= 90) 
    ]
)
filteredRentalDataFrame
```

- Simpan data terbaru tersebut ke file baru dengan nama `new_rental_data.csv` tanpa menggunakan index dari data sebelumnya

```python
filteredRentalDataFrame.to_csv('data/new_rental_data.csv', index=False)

newRentalDataFrame = pd.read_csv('data/new_rental_data.csv')
newRentalDataFrame
```
